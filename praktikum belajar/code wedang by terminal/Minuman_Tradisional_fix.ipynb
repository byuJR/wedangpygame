```python
# program_kasir.py

from minumann import Minuman, input_banyak_minuman

menu_minuman = {
    "WJ": Minuman("Wedang Jahe", 10000),
    "WU": Minuman("Wedang Uwuh", 15000),
    "WR": Minuman("Wedang Ronde", 20000),
    "BG": Minuman("Bajigur", 9000),
    "ST": Minuman("Sekoteng", 15000),
    "BD": Minuman("Bandrek", 7000),
    "STMJ": Minuman("Susu Telur Madu Jahe", 18000),
    "ET": Minuman("Es Teler", 23000),
    "ED": Minuman("Es Dawet", 12000),
}

print('''
===============================================
                PROGRAM KASIR
              MINUMAN TRADISONAL
===============================================
Kode  | Minuman               | Harga/Item
-----------------------------------------------
WJ    | Wedang Jahe           | 10000
WU    | Wedang Uwuh           | 14000
WR    | Wedang Ronde          | 20000
BG    | Bajigur               | 9000
ST    | Sekoteng              | 15000
BD    | Bandrek               | 7000
STMJ  | Susu Telur Madu Jahe  | 18000
ET    | Es Teler              | 23000
ED    | Es Dawet              | 12000

================================================
''')

ulangi_transaksi = "Y"
while ulangi_transaksi.upper() == "Y":
    total_transaksi = 0  # Menyimpan total harga dari seluruh minuman dalam satu transaksi

    ulangi = "Y"
    while ulangi.upper() == "Y":
        kode_minuman = input("Minuman yang dibeli [WJ/WU/WR/BG/ST/BD/STMJ/ET/ED]: ").upper()

        if kode_minuman in menu_minuman:
            minuman = menu_minuman[kode_minuman]
            total_harga = minuman.detail_pilihan(input_banyak_minuman())

            # Tambahkan total harga minuman ke total transaksi
            total_transaksi += total_harga

            # DIskon
            if total_transaksi > 100000:
                diskon = 0.25  # 25% discount
                discount_amount = round(total_transaksi * diskon)
                total_transaksi_after_discount = total_transaksi - discount_amount
                print(f"Selamat! Anda mendapatkan diskon 25% sebesar Rp {discount_amount}")
                print(f"Total harga sebelum diskon: Rp {total_transaksi}")

                print(f"Total harga setelah diskon: Rp {total_transaksi_after_discount}")
            else:
                discount_amount = 0
                total_transaksi_after_discount = total_transaksi
                print(f"Tidak ada diskon. Total harga: Rp {total_transaksi}")

            while True:
                try:
                    uang_bayar = int(input("Uang yang dibayar: "))
                    if uang_bayar < total_transaksi_after_discount:
                        print("Maaf, uang yang dibayarkan tidak mencukupi. Silakan bayar dengan cukup.")
                    else:
                        break
                except ValueError:
                    print("Masukkan angka")

            uang_kembali = uang_bayar - total_transaksi_after_discount

            print(f"Uang kembalian: Rp {uang_kembali}")
            print("================================================")
            ulangi = input("Tambah minuman lagi? [Y/N]: ")
            print("================================================")
        else:
            print("Kode minuman tidak valid. Minuman tidak terdaftar dalam menu. Silahkan coba lagi.")

    print(f"Total Transaksi: Rp {total_transaksi_after_discount}")
    print("=======================================")

    ulangi_transaksi = input("Ulangi transaksi? [Y/N]: ")
    print("================================================")

print("Terima Kasih atas Kunjungan Anda!")


```
